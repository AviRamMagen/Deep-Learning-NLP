```python
file_names=[
'business.txt',
'entertainment.txt',
'health.txt',
'politics.txt',
'sports.txt',
'technology.txt']
data=[]
label_to_class=[]
for i,name in enumerate(file_names):
    label_to_class.append(name)
    with open('data//'+file_names[i], 'r') as openfileobject:
        for line in openfileobject:
            data.append((line,i))

random.Random(4).shuffle(data)   
```

```python
Valitation_split=0.3
train=data[:-int(Valitation_split*len(data))]
test=data[-int(Valitation_split*len(data)):]
```

```python
train_x,train_y=zip(*train)
test_x,test_y=zip(*test)
```

```python
def print_report(pipe):
    y_test = test_y
    y_pred = pipe.predict(test_x)
    report = metrics.classification_report(y_test, y_pred,
        target_names=file_names)
    print(report)
    print("accuracy: {:0.3f}".format(metrics.accuracy_score(y_test, y_pred)))
def get_mistake(pipe):
    y_test = test_y
    y_pred = pipe.predict(test_x)
    mistakes=[]
    for i in range(len(y_test)):
        if y_test[i]!=y_pred[i]:
            mistakes.append((test_x[i],y_pred[i],y_test[i]))
    return mistakes
```

```python
vec = TfidfVectorizer()
clf = LogisticRegressionCV()
pipe = make_pipeline(vec, clf)
pipe.fit(train_x, train_y)

print_report(pipe)
```

```python
mistakes_l=get_mistake(pipe)
for i in range(40):
    display(eli5.show_prediction(clf, mistakes_l[i][0], vec=vec,
                     target_names=file_names,targets=[mistakes_l[i][1],mistakes_l[i][2]]))
                     
    print
    print
```

```python
from collections import defaultdict
from bz2 import BZ2File
import os
from datetime import datetime
import string
import wikipedia
import nltk

def get_categories(categories_filename):
    categories = defaultdict(list)
    print("Parsing the NT categories file")
    for l, line in enumerate(BZ2File(categories_filename)):
        split = line.split()
        if len(split) != 4:
            print("ignoring malformed line: " + line)
            continue
        key=split[0][29:-1]
        val=split[2][38:-1]
        categories[key].append(val)
        if l % 1000000 == 0:
            print("[%s] line: %08d" % (datetime.now().isoformat(), l))
    return dict(categories)

article_categories_filename='article_categories_en.ttl-2.bz2'
abstDict=get_categories(article_categories_filename)
```
